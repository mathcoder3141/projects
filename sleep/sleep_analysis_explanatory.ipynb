```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
pd.set_option('display.max_columns', 31)
sleep = pd.read_csv('./sleeping-alone-data.csv')
```

# Breakdown Of Education By Gender

```python
width = 0.15  # the width of the bars
labels = [gender for gender in sleep['Gender'].unique() if gender is not np.nan]

# bar heights
bachelor = [x for x in sleep[sleep["Education"] == 'Bachelor degree']['Gender'].value_counts()]
graduate = [x for x in sleep[sleep["Education"] == "Graduate degree"]['Gender'].value_counts()]
some_college = [x for x in sleep[sleep["Education"] == "Some college or Associate degree"]['Gender'].value_counts()]
high_shcool = [x for x in sleep[sleep["Education"] == "High school degree"]['Gender'].value_counts()]
no_degree =[x for x in sleep[sleep["Education"] == "Less than high school degree"]['Gender'].value_counts()]

x1 = np.arange(len(bachelor))  # the label locations
x2 = [x + width for x in x1]
x3 = [x + width for x in x2]
x4 = [x + width for x in x3]
x5 = [x + width for x in x4]

plt.bar(x1, graduate, width = width, label = 'Graduate degree')
plt.bar(x2, bachelor, width = width, label = 'Bachelor degree')
plt.bar(x3, some_college, width = width, label = 'Some College  or Assoc degree')
plt.bar(x4, high_shcool, width = width, label = 'Diploma')
plt.bar(x5, no_degree, width = width, label = 'No diploma')
plt.xticks([x + 2 * width for x in range(len(labels))], labels)
plt.xlabel('Gender', fontweight='bold')
plt.ylabel('Number of People', fontweight='bold')
plt.title('Highest Level Of Education')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
